```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://fzmovie.co.za/"
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
scraped_movies = soup.find_all('div', class_="ultp-block-content")
scraped_movies
```

```python
# parse movie names
movies = []
for movie in scraped_movies:
  title_tag = movie.find(['h2', 'h3'], class_="ultp-block-title")
  if title_tag:
    movies.append(title_tag.get_text().replace('\n',""))

display(movies)
```

```python
# store the scraped data

data = pd.DataFrame()
data['Movie Name'] = movies
data.head()
```

```python
data.to_csv('Fzmovies.csv', index=False)
```
